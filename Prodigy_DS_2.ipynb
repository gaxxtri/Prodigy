Perform data cleaning and exploratory data analysis (EDA) on a dataset of your choice, such as the Titanic dataset from Kaggle. Explore the relationships between variables and identify patterns and trends in the data.

```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv("titanic.csv")

# Display the first few rows of the dataset
titanic_data.head()
```

```python
import piplite
await piplite.install('seaborn')
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Visualization of Individual Variables
# Distribution of Age
plt.figure(figsize=(10, 6))
sns.histplot(titanic_data ['Age'], bins=30, kde=True)
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Count plot for gender
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', data=titanic_data)
plt.title('Gender Distribution of Titanic Passengers')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Count plot for class
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', data=titanic_data)
plt.title('Class Distribution of Titanic Passengers')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()

# Exploring relationships between variables
# Survival rate by gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# Survival rate by class
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic_data)
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(titanic_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
