```python
import pandas as pd

# Load the Titanic dataset
titanic_data = pd.read_csv('titanic.csv')

# Display the first few rows of the dataset
titanic_data.head()
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of genders (bar chart)
gender_counts = titanic_data['Sex'].value_counts()

plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

# Plot the distribution of ages (histogram)
plt.figure(figsize=(10, 6))
titanic_data['Age'].dropna().plot(kind='hist', bins=30, edgecolor='black')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
